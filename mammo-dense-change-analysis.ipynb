### General formatting functions used to present results

```r
###################################################
### code chunk number 2: some general functions
### for nice formatting e.g. p-vals
###################################################

fn.format<-function(ind, ndigit=2)
{
    format(round(ind,ndigit), nsmall=ndigit)
}

fn.formatp<-function(ind, ndig=3)
{
    if(length(ind)==1){
        if(ind>=0.5){
            roundstop<-1}
        else {
                roundstop<-min(ndig,ndig-sum(sapply(1:ndig, function(idx) 10^idx*ind>=1))+2)
            }
        
        myout<-fn.format(ind, roundstop)
    } else {
        roundstop<-sapply(ind, function(ind2) min(ndig,ndig-sum(sapply(1:ndig, function(idx) 10^idx*ind2>=1))+2))
        roundstop[ind>=0.5]<-1
        myout<-sapply(1:length(ind), function(idx) fn.format(ind[idx], roundstop[idx]))
    }
    
    
    t.inc<-ind < (1/(10^ndig))
    if(sum(t.inc)>0){
        myout[t.inc]<-sapply(myout[t.inc], function(ind) paste("<0",substr(ind,2,ndig+1),"1", sep=""))
    }
    
    myout
}


fn.formatci<-function(inp, inci1, inci2, ndig=2, notxt=TRUE)
{
    if(notxt){
        paste(fn.format(inp, ndig), " (", fn.format(inci1,ndig), "-", fn.format(inci2,ndig), ")", sep="")
    }
    else
    {
        paste(fn.format(inp, ndig), " (95%CI ", fn.format(inci1,ndig), "-", fn.format(inci2,ndig), ")", sep="")
    }
}


fn.formatmediqr<-function(inall, ndig=2, notxt=TRUE)
{
    if(notxt){
        paste(fn.format(inall[2], ndig), " (", fn.format(inall[1],ndig), "-", fn.format(inall[3],ndig), ")", sep="")
    }
    else
    {
        paste(fn.format(inall[2], ndig), " (IQR ", fn.format(inall[1],ndig), "-", fn.format(inall[3],ndig), ")", sep="")
    }
}

fn.formatmediqrrange<-function(inall, ndig=2)
{
    
    paste(fn.format(inall[3], ndig), " (", fn.format(inall[2],ndig), "-", fn.format(inall[4],ndig),", ", fn.format(inall[1],ndig), "-", fn.format(inall[5],ndig), ")", sep="")
    
}

fn.formatmeansd<-function(inall, ndig=2, notxt=TRUE)
{
    if(notxt){
        paste(fn.format(inall[1], ndig), " (", fn.format(inall[2],ndig), ")", sep="")
    }
    else
    {
        paste(fn.format(inall[1], ndig), " (SD ", fn.format(inall[2],ndig), ")", sep="")

    }
}

fn.formatmeansd2<-function(inall, ndig=2)
{

        paste(fn.format(inall[1], ndig), " (", fn.format(inall[2],ndig), ", ",  fn.format(inall[1]/inall[2],ndig+1), ")", sep="")

}


```

The datafile has the following columns
- `X` : Row number 
- `ID` : Woman identifier
- `Time_long_file`: Year, from -1 being one year prior to study start, to 5y
- `Tamoxifen`: Indicator (0,1) as to whether women received tamoxifen in the year
- `Tam_FHrisk_long`: categorical variable for group (FH=family history clinic, T: TAM-PREV)
- `pd`: Stratus percentage density (%)
- `densearea_cm2`: Stratus dense area (cm^2)
- `Density`: blinded visual assessment of percentage density(%)
- `DensityUnblind`: unblinded visual assessment of percentage density(%)
- `StudyDenseArea`: Densitas dense area
- `StudyBreastArea`: Densitas breast area
- `StudyBreastDensity`: Densitas percent density (%)
- `VBD`: Volpara volumetric percentage breast density (%)
- `BVcm3`: Volpara volumetric breast volume (cm^3)
- `FGVcm3`: Volpara volumetric fibroglandular volume (cm^3)
- `Age`: Age (y)
- `BMI_entry`: Body mass index (kg/m^2) at entry
- `Parity_entry`: Whether had children
- `QAgefirstchildborn`: Age at first child (y)
- `Lifetimerisk`: Projected lifetime risk of breast cancer
- `NNVAS`: NNVAS percentage density (%)

```r
###################################################################
### code chunk 4: Baseline characteristics - supplementary table 1
###################################################################

mybaseline<-mydta[mydta$"Time_long_file"==0,]

fn.quantstat<-function(ind, ndig=0){
    myout1<-fn.formatmediqrrange(quantile(ind, c(0,0.25, 0.5, 0.75,1), na.rm=TRUE),ndig)
    myout2<-sum(is.na(ind))
    myout2b<-sum(!is.na(ind))
    myout2b<-paste(myout2b, " (", fn.format(myout2/length(ind)*100,0), "%)", sep="")
    c(myout2b, myout1)
}

fn.quantstatsd<-function(ind, ndig=0){
    myout1<-fn.formatmeansd(c(mean(ind, na.rm=TRUE), sd(ind, na.rm=TRUE)), ndig)
    myout2<-sum(is.na(ind))
    myout2b<-sum(!is.na(ind))
    myout2b<-paste(myout2b, " (", fn.format(myout2/length(ind)*100,0), "%)", sep="")
    c(myout2b, myout1)
}

fn.quantstatsd2<-function(ind, ndig=0){
    myout1<-fn.formatmeansd2(c(mean(ind, na.rm=TRUE), sd(ind, na.rm=TRUE)), ndig)
    myout2<-sum(is.na(ind))
    myout2b<-sum(is.na(ind))
    myout2b<-paste(myout2b, " (", fn.format(myout2/length(ind)*100,0), "%)", sep="")
    c(myout2b, myout1)
}


fn.binstat<-function(ind){
    myout1<-sum(ind, na.rm=TRUE)
    myout1b<-paste(myout1, " (", fn.format(myout1/sum(!is.na(ind))*100,0), "%)", sep="")
    myout2<-sum(is.na(ind))
    myout2b<-sum(!is.na(ind))
    myout2b<-paste(myout2b, " (", fn.format(myout2/length(ind)*100,0), "%)", sep="")
    c(myout2b, myout1b)
}

fn.summarystat<-function(mybaseline){
    
    myagestat<-fn.quantstat(mybaseline$"Age")

    myBMIstat<-fn.quantstat(mybaseline$"BMI_entry",1)

    myTCstat<-fn.quantstat(as.double(substr(as.character(mybaseline$"Lifetimerisk"),1,4)),1)

    myageFBstat<-fn.quantstat(mybaseline$"QAgefirstchildborn",1)

    myparity<-mybaseline$Parity_entry=="Yes";         myparity[mybaseline$Parity_entry==""]<-NA

    myparitystat<-fn.binstat(myparity)

    rbind(myagestat, myBMIstat, myTCstat, myparitystat, myageFBstat)


}

mybaseline<-mydta[mydta$"Time_long_file"==0,]

mytab1a<-fn.summarystat(mybaseline[mybaseline$"Tam_FHrisk_long"=="T",])

mytab1b<-fn.summarystat(mybaseline[mybaseline$"Tam_FHrisk_long"=="FH",])
    
mylabs<-c("Age (y)", "BMI (kg/m2)", "Lifetime risk (%)", "Parous (n, %)", "Age first born (y)")
    
mytab1<-cbind(mylabs,mytab1a, mytab1b)
    
colnames(mytab1)<-c("Variable", "TP, N (% unknown)", "TP, summary*", "FH, N (% unknown)", "FH, summary*")
    
myxtab1<-xtable(mytab1, caption="Baseline characteristics of participants (analysis sample A). Footnote: *summary gives median (interquartile range, range) for continuous variables and n (\\%) for parous; TP: Tamoxifen; FH: No tamoxifen", align='llrrrr', label='tlb-1')
```

```r
###################################################
### code chunk number 10: Correlation analysis
###################################################

fn.betweencorr<-function(inx, iny)
{
    
    t.inc1<-is.na(inx)==FALSE
    t.inc2<-is.na(iny)==FALSE
    t.inc<-t.inc1*t.inc2
    t.inc[t.inc==FALSE]<-NA
    
    t.ni<-rowSums(t.inc, na.rm=T)
    
                  t.d<-cbind(rowMeans(inx*t.inc, na.rm=T),rowMeans(iny*t.inc, na.rm=T))
    
    t.d2<-t.d[t.ni>0,]
    
    t.ni2<-t.ni[t.ni>0]
    
    t.n<-length(t.ni2)
    
    t.r<-cor(rep(t.d2[,1], t.ni2), rep(t.d2[,2], t.ni2))
    
    t.ts<-sqrt(t.n-3)*(.5*log((1+t.r)/(1-t.r)))
    if(t.r>0)
    {t.p<-1-pnorm(t.ts)
    } else {t.p<-pnorm(t.ts)}
    
    c(t.r,t.p)
    
}

fn.rmeancor<-function(inx, iny, inid){
      
    t.fitdtan<-data.frame(y=iny, x=inx, id=inid)   
    
    t.lm<-lm(y~as.factor(id)+x, t.fitdtan)
    
    mycor<-sign(coef(t.lm)[length(coef(t.lm))]) * sqrt(anova(t.lm)[2,2] / sum(anova(t.lm)[2:3,2])) ## correct sign                                                                                          
    c(mycor,summary.aov(t.lm)[[1]][2,5])
    
}

fn.dencor<-function(inname1, inname2, indta){


    thisx<-indta[,which(colnames(indta)==inname1)]

    thisy<-indta[,which(colnames(indta)==inname2)]

    thisout<-fn.rmeancor(thisx, thisy, indta$ID)

    thisout

}

fn.dencor.between<-function(inname1, inname2, indta){


    thisx<-indta[,which(colnames(indta)==inname1)]

    thisy<-indta[,which(colnames(indta)==inname2)]

    thisid<-unique(indta$ID)

    thismtx<-matrix(NA, nrow=length(thisid), ncol=7)

    thismtx2<-matrix(NA, nrow=length(thisid), ncol=7)

    for(idx in 1:length(thisid)){
        for(idy in 1:7){
            thisinc<-(indta$ID==thisid[idx] & indta$"Time_long_file"==(idy-2))
            if(sum(thisinc)>0){
                thismtx[idx, idy]<-thisx[thisinc]
                thismtx2[idx, idy]<-thisy[thisinc]
                }
        }
    }


    thisout<-fn.betweencorr(thismtx, thismtx2)

    thisout

    }


fn.matrixrm<-function(myden.name, t.inc, ndig=2){

    myrmcor<-sapply(1:(length(myden.name)-1), function(idx) sapply((idx+1):length(myden.name), function(idy) fn.dencor(myden.name[idx], myden.name[idy], mydta[t.inc,])))

    myrmcor.name<-sapply(1:(length(myden.name)-1), function(idx) sapply((idx+1):length(myden.name), function(idy) paste(myden.name[idx], myden.name[idy])))

    myrmcorv<-unlist(myrmcor)

    myrmcorv.est<-fn.format(myrmcorv[seq(1,length(myrmcorv), by=2)], ndig)

    myrmcorv.p<-fn.format(-log10(myrmcorv[seq(2,length(myrmcorv), by=2)]), ndig=1)

    myout<-matrix(NA, nrow=length(myden.name),ncol=length(myden.name))

    myout[lower.tri(myout)]<-myrmcorv.est

    myoutP<-matrix(NA, nrow=length(myden.name),ncol=length(myden.name))

    myoutP[lower.tri(myoutP)]<-myrmcorv.p

    ##check  
    myout2<-myout
    myout2[lower.tri(myout2)]<-unlist(myrmcor.name)

    rownames(myout)<-myden.name;     colnames(myout)<-myden.name
    rownames(myoutP)<-myden.name;     colnames(myoutP)<-myden.name
        list(myout, myoutP) 
}


fn.matrixrm.between<-function(myden.name, t.inc, ndig=2){


    myrmcor<-sapply(1:(length(myden.name)-1), function(idx) sapply((idx+1):length(myden.name), function(idy) fn.dencor.between(myden.name[idx], myden.name[idy], mydta[t.inc,])))


    myrmcorv<-unlist(myrmcor)

    myrmcorv.est<-fn.format(myrmcorv[seq(1,length(myrmcorv), by=2)], ndig)

    myrmcorv.p<-fn.format(-log10(myrmcorv[seq(2,length(myrmcorv), by=2)]), ndig=1)

    myout<-matrix(NA, nrow=length(myden.name),ncol=length(myden.name))

    myout[lower.tri(myout)]<-myrmcorv.est

    myoutP<-matrix(NA, nrow=length(myden.name),ncol=length(myden.name))

    myoutP[lower.tri(myoutP)]<-myrmcorv.p

    rownames(myout)<-myden.name;     colnames(myout)<-myden.name
                                                                   
    rownames(myoutP)<-myden.name;     colnames(myoutP)<-myden.name
        
   list(myout, myoutP) 
                                                                   
}



myden.name<-c( "Density","DensityUnblind","pd", "StudyBreastDensity",  "VBD", "densearea_cm2", "StudyDenseArea" , "FGVcm3","StudyBreastArea",  "BVcm3", "NNVAS")

t.inc<- !is.na(mydta$VBD) & !is.na(mydta$pd) & !is.na(mydta$StudyBreastDensity) & !is.na(mydta$Density) & mydta$"Tam_FHrisk_long"!="FH"
myout1<-fn.matrixrm(myden.name, t.inc)
myout1btw<-fn.matrixrm.between(myden.name, t.inc)

myden.name<-c( "Density","pd", "StudyBreastDensity",  "VBD", "densearea_cm2", "StudyDenseArea" , "FGVcm3","StudyBreastArea",  "BVcm3", "NNVAS")
t.inc<- !is.na(mydta$VBD) & !is.na(mydta$pd) & !is.na(mydta$StudyBreastDensity) & !is.na(mydta$Density) & mydta$"Tam_FHrisk_long"=="FH"
myout2<-fn.matrixrm(myden.name, t.inc)
myout2btw<-fn.matrixrm.between(myden.name, t.inc)

                                                                                                                                   
colnames(myout1[[1]])<-c("Visual %", "Visual Unblind %", "Stratus %", "Densitas %", "Volpara %", "Stratus DA", "Densitas DA", "Volpara DV", "Densitas BA", "Volpara TV", "NNVAS %"); rownames(myout1[[1]])<-colnames(myout1[[1]])
colnames(myout1btw[[1]])<-colnames(myout1[[1]]); rownames(myout1btw[[1]])<-colnames(myout1[[1]])
colnames(myout2[[1]])<-c("Visual %", "Stratus %", "Densitas %", "Volpara %", "Stratus DA", "Densitas DA", "Volpara DV", "Densitas BA", "Volpara TV", "NNVAS %"); rownames(myout2[[1]])<-colnames(myout2[[1]])
colnames(myout2btw[[1]])<-colnames(myout2[[1]]); rownames(myout2btw[[1]])<-colnames(myout2[[1]])    

myxtab.corrrm<-xtable(myout1[[1]], caption="Repeated measures correlation (within woman) between density measures for Tamoxifen group, analysis sample (A).", align='lrrrrrrrrrrr', label='tbl:tpcor')

myxtab.corrrmBet<-xtable(myout1btw[[1]][2:(nrow(myout1btw[[1]])),1:(ncol(myout1btw[[1]])-1)], caption="Repeated measures correlation (between women) between density measures for tamoxifen group, analysis sample (A).", align='lrrrrrrrrrr', label='tbl:tpcorbtw')

    
myxtab.corrrmP<-xtable(myout1[[2]], caption="Repeated measures correlation between density measures for Tamoxifen group, -log10 P, analysis sample (A).", align='lrrrrrrrrrrr', label='tbl:tpcor2')

myxtab.corrrm.FH<-xtable(myout2[[1]], caption="Repeated measures correlation (within women) between density measures, no tamoxifen group, analysis sample (A).", align='lrrrrrrrrrr', label='tbl:tpcorFH')
    
myxtab.corrrm.FHBet<-xtable(myout2btw[[1]][2:(nrow(myout2btw[[1]])),1:(ncol(myout2btw[[1]])-1)], caption="Repeated measures correlation (between women) between density measures, no tamoxifen group, analysis sample (A).", align='lrrrrrrrrr', label='tbl:tpcorFHbtw')
    
myxtab.corrrm.FHP<-xtable(myout2[[2]], caption="Repeated measures correlation between density measures no tamoxifen group, -log10 P, analysis sample (A).", align='lrrrrrrrrrr', label='tbl:tpcorFH2')

##Supplementary Figure
postscript(file="fig-densitpairs-t.eps", paper="special",width=6.0,height=6.0,horiz=F,pointsize=10)
pairs(cbind(sqrt(mydta$Density)*10,mydta$pd, mydta$StudyBreastDensity, mydta$VBD, mydta$NNVAS), labels=c("Visual sqrt(%) x10", "Stratus sqrt(%) x 10", "Densitas sqrt(%) x 10", "Volpara ln(%) x 10", "VAS sqrt(%) x10"), cex=0.3, pch=".") 
dev.off()

pairs(cbind(sqrt(mydta$Density)*10,mydta$pd, mydta$StudyBreastDensity, mydta$VBD, mydta$NNVAS), labels=c("Visual sqrt(%) x10", "Stratus sqrt(%) x 10", "Densitas sqrt(%) x 10", "Volpara ln(%) x 10", "VAS sqrt(%) x10"), cex=0.3, pch=".") 
  
```

```r
############################a###############################
### code chunk number 13: density change cutpoint analysis
###########################################################
## How much change can we detect at year 1. 
    
t.inc<- !is.na(mydta$VBD) & !is.na(mydta$pd) & !is.na(mydta$StudyBreastDensity) & !is.na(mydta$NNVAS)

t.ex<- (mydta$Time_long_file>-1) & (mydta$Time_long_file <2) & is.na(mydta$Density)

t.inc<- t.inc & !t.ex

myid012 <- table(mydta$ID[t.inc], mydta$Time_long_file[t.inc])

idinc<-rowSums(myid012[,c(2,3,4)])==3

idincid<-names(idinc)[idinc]

mydta.012<-mydta[(mydta$ID %in% idincid) & mydta$Time_long_file<3,]

mydta.012<-mydta.012[order(mydta.012$ID, mydta.012$Time_long_file),]

ch.bl<-mydta.012[mydta.012$Time_long_file==0,]

ch.bl0<-mydta.012[mydta.012$Time_long_file== -1,]

ch.blprev <- ch.bl

ch.blprev[ch.blprev$ID %in% ch.bl0$ID,]<-ch.bl0

ch.fu1<-mydta.012[mydta.012$Time_long_file==1,]

ch.fu2<-mydta.012[mydta.012$Time_long_file==2,]

## Visual change 01
mydenchangeRuth<- -ch.bl[, which(colnames(ch.bl)=="Density")] + ch.fu1[, which(colnames(mydta)=="Density")]


## 1st plots on standardised scale
##Status, Densitas, Volpara, NNVAS; all %density
mydenname<-c("pdz","densitaspdz", "VBDz","NNVASz")
##mydenname<-c("pd","StudyBreastDensity", "VBD","NNVAS") ##if not standardised scale use this

mydenlabel<-c("Stratus PD", "Densitas PD", "Volpara PD", "NNVAS PD")

mydenchange01<-sapply(mydenname, function(inname) -ch.bl[, which(colnames(ch.bl)==inname)] + ch.fu1[, which(colnames(mydta)==inname)])

mydenchange<-sapply(mydenname, function(inname) -ch.bl[, which(colnames(ch.bl)==inname)] + ch.fu1[, which(colnames(ch.fu1)==inname)]/2 + ch.fu2[, which(colnames(ch.fu2)==inname)]/2 )

mydenchange02<-sapply(mydenname, function(inname) -ch.bl[, which(colnames(ch.bl)==inname)] + ch.fu2[, which(colnames(mydta)==inname)])

mydenchange0012<-sapply(mydenname, function(inname) 0.5*(-ch.bl[, which(colnames(ch.bl)==inname)] - ch.blprev[, which(colnames(ch.blprev)==inname)] + ch.fu1[, which(colnames(ch.fu1)==inname)] + ch.fu2[, which(colnames(ch.fu2)==inname)]) )

mydenchange001<-sapply(mydenname, function(inname) (-ch.bl[, which(colnames(ch.bl)==inname)] - ch.blprev[, which(colnames(ch.blprev)==inname)])/2 + ch.fu1[, which(colnames(ch.fu1)==inname)] )

mydenchange.arm<-ch.bl$"Tam_FHrisk_long"   

mydenchange.raw01<-data.frame(arm=rep(mydenchange.arm,4), density=c(mydenchange01[,1], mydenchange01[,2], mydenchange01[,3], mydenchange01[,4]), group=factor(rep(1:4, each=nrow(mydenchange01)), labels=mydenlabel))

mydenchange.raw<-data.frame(arm=rep(mydenchange.arm,4), density=c(mydenchange[,1], mydenchange[,2], mydenchange[,3], mydenchange[,4]), group=factor(rep(1:4, each=nrow(mydenchange)), labels=mydenlabel))

mydenchange.raw02<-data.frame(arm=rep(mydenchange.arm,4), density=c(mydenchange02[,1], mydenchange02[,2], mydenchange02[,3], mydenchange02[,4]), group=factor(rep(1:4, each=nrow(mydenchange02)), labels=mydenlabel))

mydenchange.raw0012<-data.frame(arm=rep(mydenchange.arm,4), density=c(mydenchange0012[,1], mydenchange0012[,2], mydenchange0012[,3], mydenchange0012[,4]), group=factor(rep(1:4, each=nrow(mydenchange0012)), labels=mydenlabel))

mydenchange.raw001<-data.frame(arm=rep(mydenchange.arm,4), density=c(mydenchange001[,1], mydenchange001[,2], mydenchange001[,3], mydenchange001[,4]), group=factor(rep(1:4, each=nrow(mydenchange001)), labels=mydenlabel))

mydenchange.rawRuth<-data.frame(arm=mydenchange.arm, density=mydenchangeRuth, group=rep(1, length(mydenchangeRuth)))

   
    
p3<-ggplot(mydenchange.raw0012, aes(x=group, y=density, fill=arm))+ geom_boxplot()+ylim(-1,1) + ylab("Standardised density change") + xlab("Density measure") +ggtitle("(c) Mean change over 4y {(T1 + T2)/2 - (T0 + T#)/2}") + labs(fill = "Group")

p02<-ggplot(mydenchange.raw, aes(x=group, y=density, fill=arm))+ geom_boxplot()+ylim(-1,1) + ylab("Standardised density change") + xlab("Density measure") +ggtitle("(b) Change at year 2 {T2 - T0}")  + labs(fill = "Group")

p1<-ggplot(mydenchange.raw01, aes(x=group, y=density, fill=arm))+ geom_boxplot()+ylim(-1,1) + ylab("Standardised density change") + xlab("Density measure")+ggtitle("(a) Change between baseline and year 1 (T1 - T0)")  + labs(fill = "Group")

pdf("Densitychangebox.pdf", width=20)
grid.arrange(p1,  p02, ncol=2)
dev.off()

#####################
mydenname<-c("pd","StudyBreastDensity", "VBD","NNVAS")

mydenlabel<-c("Stratus PD", "Densitas PD", "Volpara PD", "NNVAS PD")

mydenchange01<-sapply(mydenname, function(inname) -ch.bl[, which(colnames(ch.bl)==inname)] + ch.fu1[, which(colnames(mydta)==inname)])


mydenchange<-sapply(mydenname, function(inname) -ch.bl[, which(colnames(ch.bl)==inname)] + ch.fu1[, which(colnames(ch.fu1)==inname)]/2 + ch.fu2[, which(colnames(ch.fu2)==inname)]/2 )

mydenchange01<-sapply(mydenname, function(inname) -ch.bl[, which(colnames(ch.bl)==inname)] + ch.fu1[, which(colnames(mydta)==inname)])

mydenchange02<-sapply(mydenname, function(inname) -ch.bl[, which(colnames(ch.bl)==inname)] + ch.fu2[, which(colnames(mydta)==inname)])


mydenchange0012<-sapply(mydenname, function(inname) 0.5*(-ch.bl[, which(colnames(ch.bl)==inname)] - ch.blprev[, which(colnames(ch.blprev)==inname)] + ch.fu1[, which(colnames(ch.fu1)==inname)] + ch.fu2[, which(colnames(ch.fu2)==inname)]) )

mydenchange001<-sapply(mydenname, function(inname) (-ch.bl[, which(colnames(ch.bl)==inname)] - ch.blprev[, which(colnames(ch.blprev)==inname)])/2 + ch.fu1[, which(colnames(ch.fu1)==inname)] )



mydenchange.arm<-ch.bl$"Tam_FHrisk_long"   
    
mydenchange.raw<-data.frame(arm=rep(mydenchange.arm,4), density=c(mydenchange[,1], mydenchange[,2], mydenchange[,3], mydenchange[,4]), group=factor(rep(1:4, each=nrow(mydenchange)), labels=mydenlabel))

mydenchange.raw01<-data.frame(arm=rep(mydenchange.arm,4), density=c(mydenchange01[,1], mydenchange01[,2], mydenchange01[,3], mydenchange01[,4]), group=factor(rep(1:4, each=nrow(mydenchange01)), labels=mydenlabel))

mydenchange.raw02<-data.frame(arm=rep(mydenchange.arm,4), density=c(mydenchange02[,1], mydenchange02[,2], mydenchange02[,3], mydenchange02[,4]), group=factor(rep(1:4, each=nrow(mydenchange02)), labels=mydenlabel))


mydenchange.raw0012<-data.frame(arm=rep(mydenchange.arm,4), density=c(mydenchange0012[,1], mydenchange0012[,2], mydenchange0012[,3], mydenchange0012[,4]), group=factor(rep(1:4, each=nrow(mydenchange0012)), labels=mydenlabel))

mydenchange.raw001<-data.frame(arm=rep(mydenchange.arm,4), density=c(mydenchange001[,1], mydenchange001[,2], mydenchange001[,3], mydenchange001[,4]), group=factor(rep(1:4, each=nrow(mydenchange001)), labels=mydenlabel))

mydenchange.rawRuth<-data.frame(arm=mydenchange.arm, density=mydenchangeRuth, group=rep(1, length(mydenchangeRuth)))



fn.summary.denchange <- function(mydenchange.raw, mycutlow, mycuthigh, mydenlabel){

    t.inc<-mydenchange.raw$arm=="FH"
    
    fh.high.i<-sapply(1:4, function(idx) (mydenchange.raw$density[t.inc & mydenchange.raw$group==mydenlabel[idx]] <= mycuthigh[idx]))
    
    tp.high.i <- sapply(1:4, function(idx) (mydenchange.raw$density[!t.inc & mydenchange.raw$group==mydenlabel[idx]] <= mycuthigh[idx]))

    fh.low.i <- sapply(1:4, function(idx) (mydenchange.raw$density[t.inc & mydenchange.raw$group==mydenlabel[idx]] >= mycutlow[idx]))
    
    tp.low.i <- sapply(1:4, function(idx) (mydenchange.raw$density[!t.inc & mydenchange.raw$group==mydenlabel[idx]] >= mycutlow[idx]))

    ##
    fh.high<- colSums(fh.high.i)
    
    tp.high <- colSums(tp.high.i)
        
    fh.low<- colSums(fh.low.i)
    
    tp.low <- colSums(tp.low.i)
    
    nfh<-tapply(mydenchange.raw$density[t.inc], mydenchange.raw$group[t.inc], length)
    
    ntp<-tapply(mydenchange.raw$density[!t.inc], mydenchange.raw$group[!t.inc], length)
    
    lowres<-cbind(fh.low, nfh-fh.high-fh.low, fh.high)
    
    highres<-cbind(tp.low, ntp-tp.high-tp.low, tp.high)

    myproptest.high <- lapply(1:4, function(idx) prop.test(c(tp.high[idx], fh.high[idx]), c(ntp[idx], nfh[idx])) )
    
    fn.protestci<-function(ind){
        paste( round(-diff(ind[[4]])*100,0), "% (", round(100*ind[[6]][1],0), "-", round(100*ind[[6]][2],0), "%)", sep="")
    }

    diffci.tpfh.high<- sapply(myproptest.high, fn.protestci)
    
    
    lowres.tbl<-matrix(paste(lowres,paste("(", fn.format(100*prop.table(lowres,1),0), "%)", sep="")), ncol=3)
    
    highres.tbl<-matrix(paste(highres,paste("(", fn.format(100*prop.table(highres,1),0), "%)", sep="")), ncol=3)

    myhlres.tbl<-cbind(fn.format(mycutlow,1), fn.format(mycuthigh,1), lowres.tbl, highres.tbl, diffci.tpfh.high)
    
    myhlres.tbl<-myhlres.tbl[,c(1,2,3,6,4,7,5,8,9)]
    
##    colnames(myhlres.tbl)<-c("No change", "Change", "FH: not", "TP: not", "FH: unknown", "TP: unknown", "FH: change", "TP: change")
    colnames(myhlres.tbl)<-c("No change", "Change", "No change (FH)", "No change (TP)", "Intermediate (FH)", "Intermediate (TP)", "Decrease (FH)", "Decrease(TP)", "Decrease diff (95CI)")
    
    myxtab.chy2summary<-xtable(myhlres.tbl, algin='lrrrrrrrrr', caption="Analysis of cutpoints to identify women who have likely changed density, or not changed. The density change group was chosen so that approximately 5% of women not receiving tamoxifen would be identified; the no change group was chose so that approximately 5% of women receiving tamoxifen would be identified.")
    
    return(list(myxtab.chy2summary, myhlres.tbl, fh.high.i, tp.high.i, fh.low.i, tp.low.i))

}


fn.summary.denchange1 <- function(mydenchange.rawin, mycutlow, mycuthigh){

    mydenchange.raw <- mydenchange.rawin[!is.na(mydenchange.rawin$density),]
    
    t.inc<-mydenchange.raw$arm=="FH"
    
    fh.high<-sum(mydenchange.raw$density[t.inc] <= mycuthigh)
    
    tp.high<-sum(mydenchange.raw$density[!t.inc] <= mycuthigh)

    fh.low<- sum(mydenchange.raw$density[t.inc] >= mycutlow)
    
    tp.low<- sum(mydenchange.raw$density[!t.inc] >= mycutlow)
    
    nfh<-tapply(mydenchange.raw$density[t.inc], mydenchange.raw$group[t.inc], length)
    
    ntp<-tapply(mydenchange.raw$density[!t.inc], mydenchange.raw$group[!t.inc], length)
    
    lowres<-cbind(fh.low, nfh-fh.high-fh.low, fh.high)
    
    highres<-cbind(tp.low, ntp-tp.high-tp.low, tp.high)

    myproptest.high <- prop.test(c(tp.high, fh.high), c(ntp, nfh))

    
    fn.protestci<-function(ind){
        paste( round(-diff(ind[[4]])*100,0), "% (", round(100*ind[[6]][1],0), "-", round(100*ind[[6]][2],0), "%)", sep="")
    }

    diffci.tpfh.high<- fn.protestci(myproptest.high)

    
    lowres.tbl<-matrix(paste(lowres,paste("(", fn.format(100*prop.table(lowres,1),0), "%)", sep="")), ncol=3)
    
    highres.tbl<-matrix(paste(highres,paste("(", fn.format(100*prop.table(highres,1),0), "%)", sep="")), ncol=3)

    myhlres.tbl<-cbind(fn.format(mycutlow,1), fn.format(mycuthigh,1), lowres.tbl, highres.tbl, diffci.tpfh.high)
    
    myhlres.tbl<-myhlres.tbl[,c(1,2,3,6,4,7,5,8,9)]
    
    names(myhlres.tbl)<-c("No change*", "Change*", "No change (FH)", "No change (TP)", "Intermediate (FH)", "Intermediate (TP)", "Decrease (FH)", "Decrease(TP)","Decrease diff (95CI)" )
    
    return(myhlres.tbl)

}




t.inc<-mydenchange.raw01$arm=="FH"

mycuthigh<-tapply(mydenchange.raw01$density[t.inc], mydenchange.raw01$group[t.inc], function(ind) quantile(ind, c(0.15)))

mycutlow<-tapply(mydenchange.raw01$density[!t.inc], mydenchange.raw01$group[!t.inc], function(ind) quantile(ind, c(0.9)))

mycutlow<-rep(0,4)

## 01: mydenchange.raw01
## 012: mydenchange.raw
## -101mydenchange.raw001
## -1012 : mydenchange.raw0012

dechres01<- fn.summary.denchange(mydenchange.raw01, mycutlow, mycuthigh, mydenlabel)

dechres012<- fn.summary.denchange(mydenchange.raw, mycutlow, mycuthigh, mydenlabel)

dechres02<- fn.summary.denchange(mydenchange.raw02, mycutlow, mycuthigh, mydenlabel)

dechres001<- fn.summary.denchange(mydenchange.raw001, mycutlow, mycuthigh, mydenlabel)

dechres0012<- fn.summary.denchange(mydenchange.raw0012, mycutlow, mycuthigh, mydenlabel)

dechresRuth<- fn.summary.denchange1(mydenchange.rawRuth, 0, -10)

myoverallrestabl<-rbind(
    rep("", 9),
    dechresRuth, 
    rep("", 9),
    dechres01[[2]][1,],  dechres02[[2]][1,], dechres0012[[2]][1,],
    rep("", 9),
      dechres01[[2]][2,], dechres02[[2]][2,], dechres0012[[2]][2,],
    rep("", 9),
    dechres01[[2]][3,], dechres02[[2]][3,], dechres0012[[2]][3,],
    rep("", 9),
    dechres01[[2]][4,], dechres02[[2]][4,], dechres0012[[2]][4,])

## X=sqrt(1)/10 - sqrt(2)/10
## 10 * X = sqrt(1) - sqrt(2)

mytxtrep<-c("T1 - T0", "T2 - T0", "(T1 + T2)/2 - (T0 + T#)/2")

myxtab.sumchangeall <- cbind(c("(a) Visual", "T1 - T0", "(b) Stratus PD", mytxtrep, "(c) Densitas PD", mytxtrep, "(d) Volpara PD", mytxtrep, "(e) NNVAS PD", mytxtrep), myoverallrestabl)

colnames(myxtab.sumchangeall)[1]<-"Measure"        

##don't show the average measure
t.tot2<-c(6, 10, 14,18)

##myxtab.chsummaryall <- xtable(myxtab.sumchangeall[-t.tot2,], algin='lrrrrrrrrrr', caption="Number of women with a decrease in breast density, no change in density or in an intermediate group based on chosen cutpoints, using analysis sample (C). Footnote: T0: density at entry; T1: density at 1y; T2: density at 2y; FH: family-history clinic (control); TP: TAM-PREV study (received tamoxifen); no change: density change greater than or equal to this is classed as no change; change: density change less than or equal to this is classed as decrease; the intermediate category is density in between these; * for automated methods the cutpoints shown are for differences in transformed percentage density (x10), being square root for area-based methods and the natural logarithm for Volpara; Decrease diff: absolute difference in density change percentage between TP and FH.", label="tbl:cutpoints")
myxtab.chsummaryall <- xtable(myxtab.sumchangeall[-t.tot2,][,c(1,2,3,5,7,9,4,6,8,10)], algin='lrrrrrrrrrr', caption="Number of women with a decrease in breast density, no change in\
 density or in an intermediate group based on chosen cutpoints, using analysis sample (C). Footnote: T0: density at entry; T1: density at 1y; T2: density at 2y; FH: family-history \
clinic (control); TP: received tamoxifen; no change: density change greater than or equal to this is classed as no change; change: density change less than or equal to this is clas\
sed as decrease; the intermediate category is density in between these; * for automated methods the cutpoints shown are for differences in transformed percentage density (x10), bei\
ng square root for area-based methods and the natural logarithm for Volpara; Decrease diff: absolute difference in density change percentage between TP and FH.", label="tbl:cutpoin\
ts")



### do groups change and by how much
fn.sumdiscord <- function(dres1, dres2, idx1, idx2, idx3){
    
    mytab<-table(dres1[[idx1]][,idx2], dres2[[idx1]][,idx3])
    
    myout1 <- sum(mytab) - sum(diag(mytab)) ##number disagree
    
    myout1 <- mytab[2,2] ## number agree
        
    ##return( c(paste(myout1, "/", sum(mytab)-mytab[1,1],sep=""), round(myout1/sum(mytab)*100,0)))
    
    return( c(paste(myout1, "/", sum(mytab)-mytab[1,1],sep=""), round( (myout1/( sum(mytab)-mytab[1,1]) )*100, 0) ) )

}


fn.discordstat<-function(dres1, dres2){
    distab<-matrix("", nrow=4, ncol=4)
    
    for(idx in 1:4){
        for(idy in (idx):4){
            thisout <- fn.sumdiscord(dres1, dres2, 4, idx, idy)
##            distab[idy, idx] <- thisout[2]
            distab[idx, idy] <- paste(thisout[1], " (", thisout[2], "%)", sep="")
            
        }
    }
    return(distab)
}


mydiscord1 <- fn.discordstat(dechres01, dechres01)

diag(mydiscord1)<-""

mydiscord2 <- fn.discordstat(dechres02, dechres02) 

diag(mydiscord2)<-""

mydiscord3 <- fn.discordstat(dechres01, dechres02) 

mydiscord <- rbind(rep("",4), mydiscord1, rep("",4), mydiscord2, rep("",4), mydiscord3)

mydiscord<-cbind(c("(a) T1 - T0", mydenlabel, "(b) T2 - T0", mydenlabel, "(c) (T1 - T0) vs (T2 - T0)", mydenlabel), mydiscord)

colnames(mydiscord) <- c("Comparison", mydenlabel)

mydiscord <- mydiscord[-c(5,10),]

myxtab.discordance<- xtable(mydiscord, caption="Agreement in density change between pairs of density measures (n=70 in total from TAM-PREV). (a) compares between baseline (T0) and first followup at one year (T1); (b) compares change between baseline and two years later (T2); and (c) considers the concordance between change measured at one year (T0 - T1) and two years (T0 - T2), including for the same measure of breast density.", align='llrrrr', label='tbl:concordance')

mydiscord


#### multiple
mygroup01 <- (apply(dechres01[[4]], 1, function(ind) paste("G:",mydenlabel[1][ind[1]], mydenlabel[2][ind[2]], mydenlabel[3][ind[3]], mydenlabel[4][ind[4]]) ))

mygroup02 <- (apply(dechres02[[4]], 1, function(ind) paste("G:",mydenlabel[1][ind[1]], mydenlabel[2][ind[2]], mydenlabel[3][ind[3]], mydenlabel[4][ind[4]]) ))


mygrouptab01<-table(rowSums(dechres01[[4]]))

mygrouptab02<-table(rowSums(dechres02[[4]]))

mygtb01 <- paste(mygrouptab01, " (", round( mygrouptab01/sum(mygrouptab01) *100,0), "%)", sep="")

mygtb02 <- paste(mygrouptab02, " (", round( mygrouptab02/sum(mygrouptab02) *100,0), "%)", sep="")

sumagree <-cbind(rbind(mygtb01, mygtb02), rep(paste(sum(mygrouptab01), "(100%)"), 2))

colnames(sumagree)<- c(0:4, "TOTAL")
rownames(sumagree)<- c("T0 - T1", "T0 - T2")

myxtab.agree<-xtable(sumagree, caption="Number of breast density measures that agree on change above cutpoint at year 1 or year 2 compared with baseline.", align='lrrrrrr', label='tbl:denagree')

sumagree
```

### Predictive test: how well can we predict density change at y2 using data to y1?

```r
###################################################
### code chunk number 14: predictive tests
###################################################

##fit model to those with density at 0, 1 and 2
fn.fitmod3<-function(thiscolname, indta, predT=1, doEXTRAPOLATE=TRUE){
                
    thiscolidx<-which(colnames(indta)==thiscolname)
    
    fitdta<-data.frame(ID=indta$ID,y=indta[,thiscolidx], t=as.integer(indta$"Time_long_file"), trt=1*(indta$"Tam_FHrisk_long"=="T"), t1=1*(indta$"Time_long_file"==1), age=(as.double(as.character(indta$age))-42)/5, bmi=indta$"BMI_entry")
        
    fitdta$bmi[is.na(fitdta$bmi)]<-mean(fitdta$bmi[!duplicated(fitdta$ID)], na.rm=TRUE) ##impute mean BMI
    fitdta$bmi<-fitdta$bmi-25
    
    fitdta$bmi<-fitdta$bmi/5

    fitdta$t1[fitdta$trt==1 & fitdta$t>0]<-1
    
    fitdta$t1tam<-fitdta$t1*fitdta$trt

    fm01<-lmer(y ~ 1 + age + bmi  + t + t1tam + (1 +t | ID)  , data=fitdta, REML=FALSE) 
            
    newdta0<-fitdta[fitdta$t==0,]
    
    newdta1<-fitdta[fitdta$t==1,]
    
    if(doEXTRAPOLATE){
        newdta2<-newdta1; newdta2$t<-2
    }
    else {
        newdta2 <- fitdta[fitdta$t==2,]
    }
        
    mypred0 <- predict(fm01, newdta0) 
    
    mypred1 <- predict(fm01, newdta1)
    
    mypred2 <- predict(fm01, newdta2)
       
    myout<- data.frame(newdta0$ID, arm=c("FH", "T")[newdta0$trt+1], density = mypred1 - mypred0, group = rep(thiscolname, length(mypred0)), d0 = mypred0, d1 = mypred1, d2=mypred2, o0 = newdta0$y, o1=newdta1$y, o2=newdta2$y)
    
    return(myout)
    
    }

##Status, Densitas, Volpara, NNVAS; all %density
mydenname<-c("pdz","densitaspdz", "VBDz", "NNVASz")

t.inc<-(mydta.012$"Time_long_file" <2) 

mydta.mod3.01 <- lapply(mydenname, function(ind) fn.fitmod3(ind, mydta.012[t.inc,], 1, TRUE))

mydta.mod3.01m <- rbind(mydta.mod3.01[[1]], mydta.mod3.01[[2]], mydta.mod3.01[[3]], mydta.mod3.01[[4]])

t.inc<-(mydta.012$"Time_long_file" < 3) 

mydta.mod3.012 <- lapply(mydenname, function(ind) fn.fitmod3(ind, mydta.012[t.inc,], 1, FALSE))

mydta.mod3.012m <- rbind(mydta.mod3.012[[1]], mydta.mod3.012[[2]], mydta.mod3.012[[3]], mydta.mod3.012[[4]])

tempall<-cbind(mydta.mod3.01m, mydta.mod3.012m)

##colnames(tempall)
## summary data - extract relevant records
mysummary<-data.frame(ID=tempall[,1], arm=tempall[,2], group=tempall$group,y0=tempall[,8], y1=tempall[,9], y2=tempall[,20], p0=tempall[,5], p2=tempall[,7])

##observed change at 2y
mysummary$obs2 <- mysummary$y2 - mysummary$y1

## predicted change at 2y from model minus predicted y0
mysummary$pred2 <- mysummary$p2 - mysummary$p0

## observed change y1 = y0
mysummary$pastchange <- mysummary$y1 - mysummary$y0

## overall mean
mysummary$predmean<-NA

for(idarm in c("FH", "T")){
    for(idden in mydenname){
        t.this <- (mysummary$arm == idarm) & (mysummary$group == idden)
        mysummary$predmean[t.this] <- mean(mysummary$pastchange[t.this])
        }
}


fn.performance <- function(insummary, thisgroup){
    
    mysummary <- insummary[thisgroup,]
    
    thisout <- c(mean((mysummary$pastchange - mysummary$obs2)^2),
      mean((mysummary$pred2 - mysummary$obs2)^2),
      mean((mysummary$predmean - mysummary$obs2)^2) )

    return(thisout)
    
    }

fn.sel.ratio<-function(insummary, inidx){
    
    t.ref <- mean( (insummary$predmean[inidx] - insummary$obs2[inidx])^2 )
    
    t.pastchange <- mean( (insummary$pastchange[inidx] - insummary$obs2[inidx])^2 )
    
    t.modchange <- mean( (insummary$pred2[inidx] - insummary$obs2[inidx])^2 )
    
    c((t.pastchange) / (t.ref), (t.modchange) / (t.ref) )    
    
    }

fn.performance.boot <- function(insummary, thisgroup){
    
    thissummary <- insummary[thisgroup,]
    
    t.boot <- boot(thissummary, fn.sel.ratio, R=2999)
    
    myout<-lapply(1:2, function(idx) boot.ci(t.boot, index=idx, type="perc"))
    
    myout2 <- lapply(myout, function(ind) c(ind$t0, ind$percent[4:5]))
    
    myout3 <- lapply(myout2, function(ind) fn.formatci(ind[1], ind[2], ind[3], 1))
    
    return(myout3)
    
    }   

## mean squared error loss 
myperformance.res <- sapply(mydenname, function(inname) fn.performance(mysummary, mysummary$group==inname & mysummary$arm=="T"))

myperformance.res2 <- sapply(mydenname, function(inname) fn.performance.boot(mysummary, mysummary$group==inname & mysummary$arm=="T"))

myperf.test<-fn.format(myperformance.res*100, 1)

myperf.test <- rbind(matrix(paste(myperf.test[1:2,], " {", myperformance.res2, "}", sep=""), nrow=2),
                     paste(myperf.test[3,], " {REF}", sep="") )

mydenlabel<-c("Stratus PD", "Densitas PD", "Volpara PD", "NNVAS PD")

colnames(myperf.test) <- mydenlabel
rownames(myperf.test) <- c("Change Y0 - Y1", "Linear mixed model", "Mean change Y0 - Y1")

myperf.xtest <- xtable(myperf.test[c(3,1),], caption="Mean squared error (x100) in analysis sample C \\{relative performance (95\\%CI)\\} for predicting density change at two years, using mean density change at year 1 (REF=reference predictor) or observed individual density change at year 1.", align='lrrrr', label='tbl:sqerrloss')
##using mean does better than model or using most recent.

myperf.xtest                             
```
